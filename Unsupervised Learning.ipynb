# 1. K- Means Clustering

It is unsupervised machine learning algorithm

```python
# creating artificial blobs of data

from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples =400 , n_features =4 , centers=5,cluster_std = 1.8, random_state =101)
```

```python
plt.scatter(data[0][:, 0] , data[0][:,1])
```

```python
plt.scatter(data[0][:, 0] , data[0][:,1], c=data[1],cmap='plasma')
```

```python
# Training our model

from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters= 5)
```

```python
kmean.fit(data[0])
```

```python
kmean.cluster_centers_
```

```python
kmean.labels_
```

```python
# Plotting to compare our classifier's output to original data


fig, (ax1,ax2)= plt.subplots(1,2,sharey= True, figsize =(10,5))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1] , c=kmean.labels_ ,cmap= 'plasma' )

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1] , c=data[1], cmap='plasma')
```

# 2. Principal Component Analysis
It is unsupervised statistical technique to examine the interrelations among a set of variables in order to identify the underlying structure of those variables.

In this we try to determine several orthogonal lines of the best fit to dataset.

```python
# importing data set

from sklearn.datasets import load_breast_cancer
cancer= load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
# creating a dataframe 

df = pd.DataFrame(data=cancer['data'], columns = cancer['feature_names'])
```

```python
df.head()
```

```python
# standardising the data so that each feature has single unit of variance

from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler().fit(df)
```

```python
scaled_data= Scaler.transform(df)
```

```python
scaled_df= pd.DataFrame(scaled_data, columns=cancer['feature_names'])
```

```python
scaled_df.head()
```

```python
# creating PCA object

from sklearn.decomposition import PCA
```

```python
# Passing number of components wwe want o have
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
# plotting graphs for our Two principal components

plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c=cancer['target'] , cmap='rainbow')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_component = pd.DataFrame( pca.components_ ,columns=cancer['feature_names'])
```

```python
# plotting to see the correlation of Components to the differnt columns

plt.figure(figsize=(12,6))
sns.heatmap(df_component, cmap='RdBu_r')
```

# 3. Recommender System

##### Items similarity Recommendation System for movies

```python
# importing Movie-Lens Data

column_names = ['user_id' , 'item_id' , 'rating' , 'timestamp']
movi = pd.read_csv('u.data' , sep='\t',names= column_names)
```

```python
movi.head()
```

```python
# movie titles accoring to item_id

movie_names = pd.read_csv('Movie_Id_Titles')
```

```python
movie_names.head()
```

```python
# merging movies with movie_names

movi = pd.merge(movi,movie_names, on= 'item_id')
```

```python
movi.head()
```

```python
# Creating dataframe with mean ratings of movies
movi.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
# movies on total no of counts of ratings

movi.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings= pd.DataFrame(movi.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['total_rating']= pd.DataFrame(movi.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
#  creating histograms to visualise the data

ratings['total_rating'].hist(bins=100)
```

```python
sns.set_style('white')
ratings['rating'].hist(bins=100)
```

```python
# plotting relation b/w actual rating and the total ratings

sns.jointplot(x='rating' , y='total_rating' ,data= ratings , alpha= 0.5)
```

```python
# creating a matrix of movies with columns for each movie title and each row for a user
# and each cell consists of rating given by user to that movie

moviemat = movi.pivot_table(index='user_id' , columns= 'title' ,values='rating')
```

```python
moviemat.head()
```

```python
# It has lots of null values 
# most rated movies

ratings.sort_values('total_rating' , ascending= False ).head(10)
```

```python
Star_wars_user_rating = moviemat['Star Wars (1977)']
```

```python
Star_wars_user_rating.head()
```

```python
# To get coreealtion with two pandas series pairwise

similar_starwars  = moviemat.corrwith(Star_wars_user_rating )
```

```python
corr_starwars = pd.DataFrame(similar_starwars,columns= ['correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('correlation', ascending=False).head(2)
```

```python
corr_starwars = corr_starwars.join(ratings['total_rating'])
```

```python
corr_starwars.head()
```

```python
#movies recomeended for user who watched Star Wars

corr_starwars[corr_starwars['total_rating']>100].sort_values('correlation', ascending =False).head()
```

```python
liar_liar= moviemat['Liar Liar (1997)']
```

```python
similar_liar = moviemat.corrwith(liar_liar)
```

```python
corre_liar = pd.DataFrame(similar_liar, columns=['correlation'])
```

```python
corre_liar=corre_liar.join(ratings['total_rating'])
```

```python
# movies recommended for user who watached Liar Liar

corre_liar[(corre_liar['total_rating']>120) & (corre_liar['total_rating']<350)].sort_values('correlation' , ascending =False).head()
```
